```python
import pandas as pd
import pymongo
import json
import csv
import os
```

```python
# convert CSV to DataFrame
file_path = os.path.join('data','otcSpreadsheet.csv')
csvDF = pd.read_csv(file_path, encoding = 'UTF-8')
csvDF.head()
```

```python
# Set connection to mongoDB
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
db = client.otc_market_finance_data

# Drop old DB in order to create a new one
db.otc_market_listing.drop()

# JSONify the CSV to load into mongoDB
records = json.loads(csvDF.T.to_json()).values()

# Insert all CSV records into mongoDB
db.otc_market_listing.insert(records)
```
