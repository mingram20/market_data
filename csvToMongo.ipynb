```python
import pandas as pd
from pymongo import MongoClient
import json
import csv
import os
```

```python
# convert CSV to DataFrame
csvDF = pd.read_csv('otcSpreadsheet.csv', encoding = 'UTF-8')
csvDF.head()
```

```python
# Set connection to mongoDB
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
db = client.otc_markets

# JSONify the CSV to load into mongoDB
records = json.loads(csvDF.T.to_json()).values()

# Insert all CSV records into mongoDB
db.otcCSV.insert(records)
```
